```python
season_id = df1['season_id'].values
competition = df1['competition_id'].values
pairings = zip(season_id, competition)
import pandas as pd


def conc_df(pairings):
    merged_events = []

    for season_id, competition_id in pairings:
        matches = sb.matches(competition_id=competition_id, season_id=season_id)

        for match_id in matches['match_id'].values:
            df = sb.events(match_id=match_id)
            shot_events = df[df['type'] == 'Shot']

            for _, shot_event in shot_events.iterrows():
                shot_index = shot_event['index']
                shot_team = shot_event['team_id']
                prev_event_index = df[(df['index'] < shot_index) & (df['type'] != 'Ball Receipt*') & (df['team_id']==shot_team)]['index'].max()

                if pd.isnull(prev_event_index):
                    continue

                prev_event = df[df['index'] == prev_event_index].iloc[0]
                prev_event_prefixed = prev_event.add_prefix('prev_')

                merged_event = {
                    'match_id': match_id,
                    **shot_event.to_dict(),
                    **prev_event_prefixed.to_dict()
                }

                merged_events.append(merged_event)

    result_df = pd.DataFrame(merged_events)

    return result_df

dfm = conc_df(pairings)
```

```python
dfc = dfm.copy()
dfc.dropna(axis = 1, how = 'all', inplace = True)
```

```python

import pandas as pd
import numpy as np

def calculate_distances(row):
    shot_freeze_frame = row['shot_freeze_frame']
    shot_x = row['x']
    shot_y = row['y']
    
    # Initialize columns for distances
    columns = {
        'goalkeeper_dx': None,
        'goalkeeper_dy': None
    }
    teammate_dx_dy = []
    opposition_dx_dy = []

    if not isinstance(shot_freeze_frame, list):
        print(f"Unexpected data format for shot_freeze_frame: {shot_freeze_frame}")
        return pd.Series(columns)

    for entry in shot_freeze_frame:
        if not isinstance(entry, dict):
            print(f"Unexpected entry format: {entry}")
            continue

        x, y = entry['location']
        dx = x - shot_x
        dy = y - shot_y

        if entry['position']['name'] == 'Goalkeeper':
            columns['goalkeeper_dx'] = dx
            columns['goalkeeper_dy'] = dy
        else:
            if entry['teammate']:
                teammate_dx_dy.append((dx, dy))
            else:
                opposition_dx_dy.append((dx, dy))
    
    # Flatten the lists and create columns dynamically
    for idx, (dx, dy) in enumerate(teammate_dx_dy, start=1):
        columns[f'teammate_{idx}_dx'] = dx
        columns[f'teammate_{idx}_dy'] = dy
        
    for idx, (dx, dy) in enumerate(opposition_dx_dy, start=1):
        columns[f'opposition_{idx}_dx'] = dx
        columns[f'opposition_{idx}_dy'] = dy

    return pd.Series(columns)
```
